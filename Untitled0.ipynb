```python
# Function to preprocess and predict the class of the image
def predict(image):
    # Resize and preprocess the image for the model
    image = tf.image.resize(image, (IMG_HEIGHT, IMG_WIDTH))
    image = image / 255.0  # Normalize the image
    image = np.expand_dims(image, axis=0)  # Add batch dimension

    # Make predictions
    predictions = model.predict(image)
    class_index = np.argmax(predictions, axis=1)[0]

    # Assuming you have a mapping of class indices to disease names
    class_names = ['Bacterial_Pneumonia', 'Corona_virus_Disease', 'Normal', 'Tuberculosis', 'Viral_Pneumonia']  # Update with actual class names
  # Update with actual class names
    return class_names[class_index]
```

```python
IMG_HEIGHT, IMG_WIDTH = 150, 150
```

```python
# Gradio interface
iface = gr.Interface(
    fn=predict,
    inputs=gr.Image(),
    outputs="text",
    title="Lung Disease Classification",
    description="Upload a chest X-ray image to classify lung diseases."
)
```
